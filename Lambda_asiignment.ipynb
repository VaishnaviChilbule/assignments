Q Sort all the numbers in a list in descending order using lambda function



```python
l1=[34,56,78,23]
#l2=print(sorted(l1,reverse=True))
l2=sorted(l1 , key = lambda l1 : l1 , reverse=True)
print(l2)
```

```python
l = [12,3,23,43,11]
sorted_list = lambda a: sorted(a, reverse=True)
result = sorted_list(l)
print(result)
```

Q Given list of strings, sort all the strings by last character of that string. . Use lambda function and normal function both.



```python
string_list = ['was',"where","what","who","which","were","whose"]
def last_char(str1):
    return str1[-1]
print(sorted(string_list, key = last_char))
print(sorted(string_list, key = (lambda str1 : str1[-1])))
```

```python
list_of_string=['Disha','Surve','Vaishnavi','Bhagyashree','Myra']
def last_char(str1):
  return str1[-1]
print(sorted(list_of_string,key=lambda str1:str1[-1]))
```

```python
Employee_list=["ram","disha","vaishnavi","Anuja","Gautamee","raju","rishabh"]
Employee_Nlist=set(Employee_list)
def sec_char(str):
     return str[1]
print(sorted(Employee_Nlist, key = sec_char))
print(sorted(Employee_Nlist, key = lambda str : str[1]))
```

Q. Take comma separated numbers as input from the user. Split it in list of strings. Now convert every string in this list to float using map function

Q Given list of integers. Use map function to calculate square of all and store in list. Then print sum of all square values.



```python
l1=[10,20,30]
def square(n):
    return n**2
l2=list(map(square,l1))
print(l2)
res=0
for val in l2:
  res=res+val
print(res)
```

```python
l1=[10,20,30]
l2=list(map(lambda x:x**2,l1))
print(l2)
res=0
for val in l2:
  res=res+val
print(res)
```
